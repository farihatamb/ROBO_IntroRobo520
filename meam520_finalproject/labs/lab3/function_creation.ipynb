```python
import sympy
from sympy import *
from sympy import symbols
from sympy.utilities.lambdify import lambdify, lambdastr
init_printing(use_unicode=True)
```

```python
t, f, x_0, x_1 = symbols('t f x_0 x_1')
p = 1/f
```

```python
position =  2*abs((t/p) - floor((t/p) + 0.5)) #1 - cos(t*f*2*pi)
position
```

```python
xpos = (1-position)*x_0 + position*x_1
xpos
```

```python
vels = diff(xpos, t)
vels
```

```python
lambdastr([f,x_0,x_1,t,], vels)
```

```python
t, f, ry, rz = symbols('t f r_y r_z')
x0 = MatrixSymbol('x_0', 3, 1)
```

```python
rads = (t * f) + (3*pi/2)
rads
```

```python
pos = x0 + Matrix([0, ry*cos(rads), (rz*sin(rads))+rz])
pos
```

```python
vel = simplify(diff(pos, t))
vel
```

```python
lambdastr([t, ry, rz, rz], vel)
```
