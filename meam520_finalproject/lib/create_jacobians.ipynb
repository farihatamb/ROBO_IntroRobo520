```python
import sympy
from sympy import *
from sympy import symbols
from sympy.utilities.lambdify import lambdify, lambdastr
init_printing(use_unicode=True)
```

```python
Q = Array(symbols(['q_1', 'q_2', 'q_3', 'q_4', 'q_5', 'q_6', 'q_7']))
Q
```

```python
d_1, d_3, d_5, d_7 = symbols('d_1 d_3 d_5 d_7')
a_3, a_4, a_6 = symbols('a_3 a_4 a_6')
```

```python
def dh_matrix(a, alpha, d, theta):
    M = Matrix([
        [cos(theta), -sin(theta)*cos(alpha), sin(theta)*sin(alpha), a*cos(theta)],
        [sin(theta), cos(theta)*cos(alpha), -cos(theta)*sin(alpha), a*sin(theta)],
        [0, sin(alpha), cos(alpha), d],
        [0, 0, 0, 1]
    ])
    return M
```

```python
T01 = dh_matrix(0, -pi/2, d_1, Q[0])
T01
```

```python
T12 = dh_matrix(0, pi/2, 0, Q[1])
T12
```

```python
T23 = dh_matrix(a_3, pi/2, d_3, Q[2])
T23
```

```python
T34 = dh_matrix(-a_4, -pi/2, 0, Q[3])
T34
```

```python
T45 = dh_matrix(0, pi/2, d_5, Q[4])
T45
```

```python
T56 = dh_matrix(a_6, pi/2, 0, Q[5])
T56
```

```python
T67 = dh_matrix(0, 0, d_7, Q[6])
T67
```

```python
components = list()
Ts = [T01, T01@T12, T01@T12@T23, T01@T12@T23@T34, T01@T12@T23@T34@T45, T01@T12@T23@T34@T45@T56, T01@T12@T23@T34@T45@T56@T67]
for i in range(7):
    components.append(diff(Ts[i][:3,:3],Q[i])@Ts[i][:3,:3].T@Ts[i][:3,:3]@Matrix([0,0,1]))
components
```

```python
J = zeros(6,7)
TT= [T01, T01@T12, T01@T12@T23, T01@T12@T23@T34, T01@T12@T23@T34@T45, T01@T12@T23@T34@T45@T56, T01@T12@T23@T34@T45@T56@T67]

r=Matrix([0,0,1])
d0x=Matrix([0,0,.141])
Rstuff=Matrix([[0,-r[2],r[1]],
                [r[2],0,-r[0]],
                [-r[1],r[0],0]])
d=Matrix([[TT[-1][0]-d0x[0]],
            [TT[-1][1]-d0x[1]],
            [TT[-1][2]-d0x[2]]])
Jvx=Rstuff@d
J[0,0]=Jvx[0]
J[1,0]=Jvx[1]
J[2,0]=Jvx[2]
J[3,0]=r[0]
J[4,0]=r[1]
J[5,0]=r[2]

for x in range(1,7):
    r=TT[x-1][:,2]
    d0x=TT[x-1][:,3]
    Rstuff=Matrix([[0,-r[2],r[1]],
                    [r[2],0,-r[0]],
                    [-r[1],r[0],0]])
    d=Matrix([[TT[-1][0]-d0x[0]],
                [TT[-1][1]-d0x[1]],
                [TT[-1][2]-d0x[2]]])
    Jvx=Rstuff@d
    J[0,x]=Jvx[0]
    J[1,x]=Jvx[1]
    J[2,x]=Jvx[2]
    J[3,x]=r[0]
    J[4,x]=r[1]
    J[5,x]=r[2]
```
